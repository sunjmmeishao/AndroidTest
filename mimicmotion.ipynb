```
def install(package):
    import sys
    !{sys.executable} -m pip install {package}

# 列表包含需要检查并可能需要安装的库及其安装命令
libraries = [
    "torch torchvision torchaudio --extra-index-url https://download.pytorch.org/whl/cu118",
    "numpy",
    "omegaconf",
    "diffusers",
    "transformers",
    "einops",
    "safetensors"
]

for lib in libraries:
    try:
        # 尝试导入库
        if ' ' in lib:  # 如果是复合命令，如 torch torchvision torchaudio ...
            exec(f"import {lib.split()[0]}")
        else:
            exec(f"import {lib}")
        print(f"{lib.split()[0] if ' ' in lib else lib} 已安装并可用.")
    except ImportError:
        print(f"{lib.split()[0] if ' ' in lib else lib} 未找到，正在安装...")
        install(lib)
        try:
            # 再次尝试导入库
            if ' ' in lib:
                exec(f"import {lib.split()[0]}")
            else:
                exec(f"import {lib}")
            print(f"{lib.split()[0] if ' ' in lib else lib} 已成功安装并导入.")
        except ImportError:
            print(f"警告：未能成功安装或导入 {lib.split()[0] if ' ' in lib else lib}.")
```

```
import requests

def get_ngrok_url():
    try:
        res = requests.get("http://localhost:4040/api/tunnels").json()
        if res["tunnels"]:
            return res["tunnels"][0]["public_url"]
    except Exception as e:
        print("❌ 获取失败:", str(e))
    return None

url = get_ngrok_url()
if url:
    print(f"\n🌐 你的公网地址是：{url}")
else:
    print("⚠️ 请确保 ngrok 已启动")
```
